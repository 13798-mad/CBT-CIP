```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("Unemployment in India - Unemployment in India.csv")
data.head(5)
```

```python
data.info()
```

```python
data.describe()
```

```python
plt.figure(figsize=(12,5))
sns.lineplot(x="Date",y="Estimated Unemployment Rate (%)",data=data,color="green",marker="o")
plt.xticks(rotation=90)
plt.show()
```

```python
region_avg_unemployment = data.groupby('Region')['Estimated Unemployment Rate (%)'].mean().reset_index()

plt.figure(figsize=(12,6))
sns.barplot(x='Region', y='Estimated Unemployment Rate (%)', data=region_avg_unemployment, palette='Set2')
plt.title('Average Unemployment Rate per Region', fontsize=15)
plt.xlabel('Region', fontsize=12)
plt.ylabel('Average Unemployment Rate (%)', fontsize=12)
plt.xticks(rotation=90)
plt.tight_layout() 
plt.show()
```

```python
region_avg_unemployment = data.groupby('Region')['Estimated Labour Participation Rate (%)'].mean().reset_index()

plt.figure(figsize=(12,6))
sns.barplot(x='Region', y='Estimated Labour Participation Rate (%)', data=region_avg_unemployment, palette='viridis')
plt.title('Average Estimated Labour Participation Rate (%)', fontsize=15)
plt.xlabel('Region', fontsize=12)
plt.ylabel('Average Estimated Labour Participation Rate (%)', fontsize=12)
plt.xticks(rotation=90)
plt.tight_layout() 
plt.show()
```

```python
region_avg_unemployment = data.groupby('Date')['Estimated Unemployment Rate (%)'].mean().reset_index()

plt.figure(figsize=(12,6))
sns.barplot(y='Date', x='Estimated Unemployment Rate (%)', data=region_avg_unemployment, palette='Set1')
plt.title('Average Unemployment Rate per Date', fontsize=15)
plt.ylabel('Date', fontsize=12)
plt.xlabel('Average Unemployment Rate (%)', fontsize=12)
plt.xticks(rotation=90)
plt.tight_layout() 
plt.show()
```
